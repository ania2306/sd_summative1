To answer the question above we need to add two new columns to the existing dataframe with the Student Overall Attendance variable and Averge Module Attendance Variable

````python
#computing each student's overall attendance rate and calling it overall attendance
overall = df.groupby("Person Code")["Attended"].mean().rename("Overall Attendance")
overall.head()
````

````python
#merging the new overall attendace rate with the main dataframe - aligined with Person Code colum
df2 = df.merge(overall, on = "Person Code")
df2.head()
````

````python
#computing average overall attendance for each module for students that attended the module
module_avg_overall = (
    df2[df2["Attended"] == 1]
    .groupby("Module Name")["Overall Attendance"]
    .mean()
)

module_avg_overall.head()
````

````python
#naming the avergane module attendance series before adding it to my dataframe
att_mean = att_mean.rename("Average Module Attendance")

#merging the average module attendace with the main dataframe
df2 = df2.merge(att_mean, on = "Module Name")
df2
````

````python
#restricting the values to cases when the studed has actually attended the module
df_att = df2[df2["Student Overall Attendance"] != 0]
````

I am going to use .corr to comput the correlation instead of scipystats pearsonr due to my pandas and numpy consistently breaking and not being able to upload the necessary module.

````python
#computing the correlation
corr = df_att["Student Overall Attendance"].corr(df_att["Average Module Attendance"])
print("Correlation:", corr)
````

Since the correlation is 0.52 (positive - both variables increase together across the dataset) that indicates that students with high attendance tend to attend modules with high average attendace. 

Simultaneously that shows that students with low attendance tend to attend low-attendance modules. 

Below I am going to plot a scatterplot using lmplot (a linear model plot) to visualise the relationship between students's attendance and average mdoule attendance.

````python
#visualising the correlation with a scatter plot
sns.lmplot(
    data = df_att,
    x = "Student Overall Attendance",
    y = "Average Module Attendance"
)

plt.title("The relationship between student overall attendance and average module attendance")
````

To answer the question using linear regression I am going to imposer relevant statsmodels module and use the same variables (renames as x and y for the sake of brevity) to run an OLS model.

````python
#linear regression answer

import statsmodels.formula.api as smf
````

````python
#computing linear regression model

#setting variables x and y since their names contain spaces
x = df_att["Student Overall Attendance"]
y = df_att["Average Module Attendance"]

model = smf.ols("x ~ y", data = df_att).fit()
print(model.summary())
````

In the resulted model of Student Overall Attendance ~ Module Average Attendance we can read that:

The R-squared value explains that 27.1% of student overall attendance can be explained by the changes in average module attendance (the model explains 27.1% of change in student overall attendance).

The low p value (p = 0.000) indicates that 0% of the model can be explained by chance.

The y coefficient (y = 0.5053) shows that a change of one unit in the average modul attendance will result in a 0.6053 point change in the student attendance rate. 

The linear regression model also confirm the positive relationship between averge module attendance and student's overall atendance - the lower mean module attendance the lower individual student mean attendance will be.
